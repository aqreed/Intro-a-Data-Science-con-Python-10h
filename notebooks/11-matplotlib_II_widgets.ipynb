## Interactuar con el código en Jupyter Notebook

Por ejemplo, sea una función que imprime un texto cualquiera rodeado por el caracter que queramos:

```python
def decorate_text(text, decorator, n):
    print(decorator*n + text + decorator*n)

decorate_text('hola', '*', 5)
```

Para cambiar el texto 'hola' y el decorador '*', habría que modificar la celda anterior y reejecutarla cuantas veces hagamos modificaciones. Para evitar este proceso, existe la función de interacción [`interact`](http://ipywidgets.readthedocs.io/en/latest/examples/Using%20Interact.html) de la librería ipywidgets:

```python
from ipywidgets import interact

interact(decorate_text, text='hola', decorator=['*', '_', '-'], n=(1, 10, 1))
```

Si en vez de la función "decorate_text" usamos otra que haga más cosas (por ejemplo plotea una gráfica), también será posible interactuar con ella.

** Ejemplo **

Una función que represente gráficamente esta expresión:

$$\sin(2 \pi f_1 t) + \sin(2 \pi f_2 t)$$

Siendo $f_1$ y $f_2$ argumentos de entrada (por defecto $10$ y $100$) y $t \in [0, 0.5]$. Además usará algún estilo de los disponibles y debe mostrar:

* Leyenda
* Título "Dos frecuencias"
* Eje x "Tiempo ($t$)"

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(a=1.0):
    x = np.linspace(-1, 3, 100)
    funcion = np.exp(-x ** (2.0*a))
    with plt.style.context("ggplot"):
        plt.plot(x, funcion, label="Func a = %s" %a)
        plt.xlabel("x")
        plt.title("Funcion interactuable")
        plt.legend()

f()
```

```python
from ipywidgets import FloatSlider, interact_manual

#interact_manual(f, a=FloatSlider(min=1, max=25, step=1))
interact(f, a=(1, 20, 1), continuous_update=False)
```
